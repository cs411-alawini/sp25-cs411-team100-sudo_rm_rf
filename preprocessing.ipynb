```python
ddid_df = pd.read_csv("Interaction Information.csv", index_col = False)

# Drop duplicate rows because there are sometimes multiple copies of the same drug-food 
# interaction but with the drug having different forms (e.g pill and liquid forms have two seperate rows)
ddid_inter = ddid_df[["Drug_ID", "Drug_Name", "Food_Herb_ID", "Food_Herb_Name", "Effect"]].drop_duplicates().reset_index().drop("index", axis = 1)
ddid_inter = ddid_inter.rename(columns = {"Drug_ID":"drug_1_rxnorn_id", "Drug_Name":"drug_1_concept_name", 
                                    "Food_Herb_ID":"drug_2_rxnorm_id", "Food_Herb_Name":"drug_2_concept_name"})
```

```python
#import pandas as pd
#import numpy as np

column_names = ["RXCUI", "LAT", "TS", "LUI", "STT", "SUI", "ISPREF", "RXAUI", 
                "SAUI", "SCUI", "SDUI", "SAB", "TTY", "CODE", "STR", "SRL", "SUPPRESS", "CVF"]
rxcui_df = pd.read_csv("RXNCONSO.RRF", delimiter = "|", names = column_names, index_col = False)
rxcui_df.head()
```

```python
# Convert to standardized Rxnorm id scheme
rxnorm_ids = [] 
not_found_drugs = []
rxcui_lowercase = rxcui_df["STR"].str.lower()
i = 0
for drug_name in ddid_inter["drug_1_concept_name"]:
    row = rxcui_df[rxcui_lowercase.isin([drug_name.lower()])]
    if row.empty:
        not_found_drugs.append(drug_name)
        rxnorm_ids.append("")
    else:
        rxnorm_ids.append(row["RXCUI"].iloc[0])
    i += 1
    if i % 1000 == 0:
        print("Iter:", i)
    #input(rxnorm_ids)
#ddid_inter["drug_1_concept_name"]
```

```python
not_found_drugs
```

```python
ddid_inter["drug_1_rxnorn_id"] = rxnorm_ids

ddid_inter["mean_reporting_frequency"] = 0.0
ddid_inter.loc[ddid_inter["Effect"] == "Possible", "mean_reporting_frequency"] = 0.5
ddid_inter.loc[ddid_inter["Effect"] == "Positive", "mean_reporting_frequency"] = 1
ddid_inter_final = ddid_inter.drop("Effect", axis = 1)
```

```python
ddid_inter_final = ddid_inter_final.drop_duplicates(["drug_1_rxnorn_id", "drug_2_rxnorm_id"])
```

```python
cols = pd.DataFrame(columns = next(df2sides_iter).columns)
ddid_2sides_format = pd.concat([cols, ddid_inter_final])
ddid_2sides_format
```

```python
ddid_2sides_format.to_csv("TWOSIDES_appended.csv", mode = "a", header = False, index = False)
df2sides_iter.close()
```

```python
unique_foods = ddid_df[["Food_Herb_ID", "Food_Herb_Name"]].drop_duplicates().reset_index()
unique_foods["SUPPRESS"] = "N"
unique_foods.rename(columns = {"Food_Herb_ID" : "RXCUI", "Food_Herb_Name" : "STR", "index" : "RXAUI"}, inplace = True)
unique_foods["RXAUI"] = unique_foods["RXAUI"] * -1


combined_ids = pd.concat((rxcui_df, unique_foods), ignore_index = True)
combined_ids.to_csv("Drug+Food IDs.csv", index = False)
```
